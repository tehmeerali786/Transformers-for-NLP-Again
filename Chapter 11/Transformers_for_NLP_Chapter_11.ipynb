```python
from transformers import pipeline
```

```python
def classify(sequence, M):
  #DistilBERTForSequenceClassification(default, model)
  nlp_cls = pipeline('sentiment-analysis')
  if M == 1:
    print(nlp_cls.model.config)
  return nlp_cls(sequence)
```

```python
seq = 1
if seq == 1:
  sequence = "The battery on my Model9x phone doesn't last more than 6 hours and I'm unhappy about that."
if seq == 2:
  sequence = "The battery on my Model9x phone doesn't last more than 6 hours and I'm unhappy about that. I was really mad! I bought a Model10x and things seem to be better. I'm supper sattisfied now."
if seq == 3:
  sequence = "The customer was unhappy."
if seq == 4:
  sequence = "The customer was unsatisfied."

print(sequence)
# display model configuration=1, default=0
M = 1
CS=classify(sequence, M)
print(CS)
```
